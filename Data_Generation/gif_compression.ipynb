```python
import imageio.v2 as imageio
import numpy as np

input_path = "vorticity_evolution.gif"
output_path = "vorticity_evolution_compressed.gif"

import imageio.v2 as imageio
import numpy as np
from PIL import Image

reader = imageio.get_reader(input_path)

frames = []
target_size = None

for i, frame in enumerate(reader):
    # Skip every other frame to reduce size
    if i % 2 != 0:
        continue

    # Convert numpy frame -> PIL image so we can safely standardize it
    img = Image.fromarray(frame).convert("RGB")

    # Set a common target size from the first kept frame
    if target_size is None:
        target_size = (img.width // 2, img.height // 2)

    # Resize every frame to exactly the same size
    img = img.resize(target_size, Image.LANCZOS)

    # Convert back to numpy
    arr = np.array(img)

    # Reduce colors a bit
    arr = (arr // 32) * 32

    frames.append(arr.astype(np.uint8))

# Save compressed GIF
imageio.mimsave(output_path, frames, duration=0.05)
```
